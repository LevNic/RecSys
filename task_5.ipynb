2) Что используют в качестве признаков?

LightFM (теги): модель LightFM использует только тегвозможности.  
LightFM (теги + идентификаторы): модель LightFM с использованиемкак функции индикатора тегов, так и функции индикатора элементов.  
LightFM (теги + о компании): модель LightFM с использованиемкак элемент, так и пользовательские функции.
